```python
# THOOTB 最小可行模擬版 v0.2（Chin Sook Ling 專用加強版）
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import random

# 常數
DAILY_TH = 1000
人口 = 100

# 公民類別
class 公民:
    def __init__(self, id):
        self.id = id
        self.餘額 = 0
        self.Maslow等級 = random.randint(2,5)
        self.最後發錢日 = None

# 公司類別
class 公司:
    def __init__(self, 名稱, 是否通過RDR=True, 囤貨傾向=0):
        self.名稱 = 名稱
        self.是否通過RDR = 是否通過RDR
        self.囤貨傾向 = 囤貨傾向  # 0=誠實, 1=超愛囤
        self.售價 = 50

# 建立世界
公民們 = [公民(f"C{i:04d}") for i in range(人口)]
公司們 = [
    公司("誠實農場", True, 0),
    公司("綠色工廠", True, 0),
    公司("囤貨奸商", False, 1),
    公司("半誠實商店", True, 0.6)
]

交易紀錄 = []
每日總餘額 = []

# 模擬 100 天
for 日 in range(1, 101):
    print(f"\n=== 第 {日} 天 ===")

    # 每日發錢
    for 人 in 公民們:
        人.餘額 = DAILY_TH
    print(f"已發放 TH{DAILY_TH} × {人口} 人 = TH{DAILY_TH*人口:,}")

    # 消費階段
    今天交易 = 0
    for 人 in 公民們:
        if random.random() < 0.75 and 人.餘額 >= 50:  # 75% 機率消費
            選的公司 = random.choice(公司們)
            if not 選的公司.是否通過RDR:
                print(f"  ❌ {人.id} 想買 {選的公司.名稱} → 被 R&D+D&R 擋下！")
                continue
            人.餘額 -= 選的公司.售價
            今天交易 += 選的公司.售價
            交易紀錄.append({"日":日, "公民":人.id, "公司":選的公司.名稱, "花費":選的公司.售價})

    # 晚上歸零
    當天剩餘 = sum(人.餘額 for 人 in 公民們)
    每日總餘額.append(當天剩餘)
    for 人 in 公民們:
        人.餘額 = 0

    print(f"今天流通 TH{今天交易:,}，晚上剩餘 TH{當天剩餘:,} → 全數歸零回收")

    # 第50天加入大亂：奸商試圖囤貨
    if 日 == 50:
        print("   ⚠️ 外部危機：囤貨奸商試圖控制 45% 物資 → 即將觸發 Pareto 20% 警報！")

print("\n🎉 100天模擬完成！系統穩定、無通貨膨脹、無人能囤積！")
df = pd.DataFrame(交易紀錄)
print(f"總交易筆數：{len(df):,} 筆")

# 畫圖證明系統健康
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(每日總餘額)
plt.title("每天結束前剩餘 TH（應該接近 0）")
plt.subplot(1,2,2)
df.groupby('日').size().plot(kind='line')
plt.title("每日交易量（越穩定越好）")
plt.show()
```

```python
# THOOTB v0.3 - 正式加入 Pareto 20% + 四大核心原則自動監控（Chin Sook Ling 專用）
import pandas as pd
import matplotlib.pyplot as plt
import random
from datetime import datetime

人口 = 200
DAILY_TH = 1000
總資源單位 = 10000  # 假設整個 zone 總共有 10000 單位物資（食物、工具等）

class 公民:
    def __init__(self, id):
        self.id = id
        self.餘額 = 0
        self.貢獻分 = 0
        self.Maslow = random.randint(2,5)

class 公司:
    def __init__(self, 名稱, 是否通過RDR=True, 囤貨傾向=0):
        self.名稱 = 名稱
        self.是否通過RDR = 是否通過RDR
        self.囤貨傾向 = 囤貨傾向
        self.持有資源 = random.randint(50, 300)  # 一開始隨機持有資源
        self.售價 = random.randint(45, 60)

# 建立世界
公民們 = [公民(f"C{i:04d}") for i in range(人口)]
公司們 = [
    公司("誠實農場", True, 0),
    公司("循環工廠", True, 0),
    公司("囤貨大王", False, 1),     # 惡意玩家
    公司("半誠實超市", True, 0.7)
]

交易紀錄 = []
每日剩餘 = []
pareto違規記錄 = []

# 模擬 365 天
for 日 in range(1, 366):
    # --- 原則一：每日發放 + 歸零 ---
    for 人 in 公民們:
        人.餘額 = DAILY_TH
    當天剩餘 = 0

    # --- 消費階段 ---
    for 人 in 公民們:
        if random.random() < 0.8 and 人.餘額 >= 50:
            選的公司 = random.choice(公司們)
            if not 選的公司.是否通過RDR:
                continue
            人.餘額 -= 選的公司.售價
            選的公司.持有資源 += 1  # 賣出去就增加持有（模擬囤貨）
            交易紀錄.append({"日":日, "公司":選的公司.名稱, "售價":選的公司.售價})

    # --- 原則三：Pareto 20% 自動偵測（白皮書最重要一刀！）---
    for 公司 in 公司們:
        佔比 = 公司.持有資源 / 總資源單位
        if 佔比 > 0.20:  # 超過 20% 就觸發警報
            pareto違規記錄.append({"日":日, "公司":公司.名稱, "佔比":round(佔比,3)})
            print(f"🚨 第{日}天 Pareto 20% 警報！{公司.名稱} 持有 {佔比:.1%} 資源 → 自動標記並啟動社區審議")

    # --- 每日歸零 ---
    當天剩餘 = sum(人.餘額 for 人 in 公民們)
    每日剩餘.append(當天剩餘)
    for 人 in 公民們:
        人.餘額 = 0

print(f"\n🎉 365 天模擬結束！共偵測到 {len(pareto違規記錄)} 次 Pareto 20% 違規")
if pareto違規記錄:
    df警報 = pd.DataFrame(pareto違規記錄)
    print(df警報.tail(10))

# 畫圖
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.plot(每日剩餘)
plt.title("每日歸零前剩餘 TH（證明無通脹）")

plt.subplot(1,3,2)
資源佔比 = [c.持有資源 for c in 公司們]
plt.bar([c.名稱 for c in 公司們], 資源佔比)
plt.axhline(總資源單位*0.2, color='red', linestyle='--', label='Pareto 20% 紅線')
plt.title("最終資源分配（紅線＝20%）")
plt.legend()

plt.subplot(1,3,3)
pd.DataFrame(交易紀錄).groupby('日').size().plot()
plt.title("每日交易量（經濟活躍度）")
plt.show()
```
